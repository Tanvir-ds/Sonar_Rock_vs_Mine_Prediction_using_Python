```python
#importing dependences
import numpy as np
import pandas as pd
```

```python
# loading the datasets
sonar_data =pd.read_csv("sonar data.csv",header=None)
sonar_data.head(30)
```

```python
# Shape of the datasets
sonar_data.shape
```

```python
# Statistical information of the datasets
sonar_data.describe()
```

```python
# Printing data types of each columns of the datasets
sonar_data.info()
```

```python
# Printing how many Rock and Mine has in 60th/target column
sonar_data[60].value_counts()
```

```python
# Statistical Mean for target columns
sonar_data.groupby(60).mean()
```

```python
# Finding missing values
sonar_data.isnull().sum()
```

```python
# Finding Duplicate values
sonar_data.duplicated().sum()
```

```python
# Separating the dataset into Dependent and Independent variables
X = sonar_data.drop(columns=[60],axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

```python
# splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
# Fitting the Training datasets to Logistic Regression model
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
# Evaluate the training datasets model
from sklearn.metrics import accuracy_score
x_train_prediction = log_reg.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("Accuracy on Train Datasets:",training_data_accuracy)
```

```python
# Evaluate the test datasets
x_test_prediction = log_reg.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("Accuracy on Test Datasets:",test_data_accuracy)
```

```python
# creating Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(x_test_prediction,y_test)
cm
```

```python
# Visualizing confusion matrix
import seaborn as sns
sns.heatmap(cm,annot=True)
```
